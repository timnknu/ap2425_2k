```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.arange(20, 400, 10)
x
```

```python
a = (1+1/x)**x
print(np.vstack((x, a)).T)
```

```python
plt.plot(x, a, '.')

plt.axhline(np.exp(1), linestyle=':')
plt.ylabel('$a_n$', fontsize='x-large')
plt.xlabel('$n$', fontsize='x-large')
plt.show()
```

```python
import scipy.optimize as sp_opt
import math

def func(x):
    return math.cos(-x) + x

sol = sp_opt.fsolve(func, [0])
print(sol)
```

```python
np.cos(sol)
```

```python
import scipy.integrate as sp_int

def rhs(y, t):
    return -2*y

t = np.linspace(0, 2, 100)
sol = sp_int.odeint(rhs, [3.0], t)
plt.plot(t, sol[:, 0])
```

```python
import re

text = """
Hello, my email is john.doe@example.com. You can also reach me at
johndoe123@gmail.com or j.doe@company-name.co.uk. Our support email
is support@ourwebsite.org.
"""

email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'

emails = re.findall(email_pattern, text)

for email in emails:
    print(email)
```

```python
def replace_at_symbol(match):
    #email = match.group(0)
    #return email.replace('@', '[SYMBOL]')
    return '[приватна інформація]'

modified_text = re.sub(email_pattern, replace_at_symbol, text)

print(modified_text)
```

```python
import urllib.request
import json

mlist = [
    {'role': 'system', 'content': 'You are a helpful assistant'},
    #{'role': 'user', 'content': 'Напиши теорему Піфагора'}
    {'role': 'user', 'content': 'Котра година в Індії?'}
]

req = { 'model': 'gpt-3.5-turbo-0125', 'messages': mlist, 'temperature': 1.5}

web_req = urllib.request.Request('https://api.openai.com/v1/chat/completions',
                             method = 'POST',
                             data = json.dumps(req).encode('utf-8'),
                             )
web_req.add_header("Content-Type", "application/json")
web_req.add_header("Authorization", "Bearer sk-proj-.....")  # сюди потрібно підставити особистий ключ

resp = urllib.request.urlopen(web_req)
```

```python
resp_text = resp.read().decode('utf-8')
resp_text
```

```python
resp_obj = json.loads(resp_text)
# resp_obj
print(resp_obj['choices'][0]['message']['content'])
```
